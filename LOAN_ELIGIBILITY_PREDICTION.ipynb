```python
# Select numeric features for correlation analysis
numeric_features = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']

# Create a correlation matrix
corr_matrix = Train_dataset[numeric_features + ['Loan_Status']].corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', annot_kws={"size": 10})
plt.title('Correlation Matrix')
plt.show()
```

corr() function calculates the correlation coefficients between these variables.

### Feature Selection using Correlation Analysis

```python
# Select numeric features for correlation analysis
numeric_features = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']

# Calculate correlation matrix
corr_matrix = Train_dataset[numeric_features + ['Loan_Status']].corr()

# Print correlation matrix
print(corr_matrix)

# Identify important features based on correlation coefficients
important_features = corr_matrix['Loan_Status'].sort_values(ascending=False)
print(important_features)
```

Feature Selection: Identify important features that have a significant impact on loan eligibility. You can use techniques like correlation analysis or feature importance from algorithms like Random Forest. 
Above code calculates the correlation matrix for numeric features and the target variable ('Loan_Status'). The important_features variable will contain the correlation coefficients, indicating the relationship strength between each feature and loan eligibility. Higher absolute values (closer to 1 or -1) suggest stronger relationships.

# BUILDING THE MODEL 

### DecisionTree

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
DTClassifier=DecisionTreeClassifier(criterion='entropy',random_state=42)
DTClassifier.fit(X_train,y_train)
```

```python
y_predicted=DTClassifier.predict(X_test)
```

```python
y_predicted
```

```python
from sklearn import metrics
print('The accuracy of decision tree is: ', metrics.accuracy_score(y_predicted,y_test))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

cm = confusion_matrix(y_test, y_predicted)
print("Confusion Matrix:")
print(cm)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_predicted)
print(f"Accuracy: {accuracy:.2f}")

# Calculate precision
precision = precision_score(y_test, y_predicted)
print(f"Precision: {precision:.2f}")

# Calculate recall
recall = recall_score(y_test, y_predicted)
print(f"Recall: {recall:.2f}")

# Calculate F1 score
f1 = f1_score(y_test, y_predicted)
print(f"F1 Score: {f1:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)

# Plot confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', annot_kws={"size": 14})

plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix for Decision Tree')
plt.show()
```
